```python
import os
import re
import pandas as pd
print(os.getcwd())
```

```python
with open('..\\datasets\\clinical_cases\\clinical_few_shot_prompt.md', 'r', encoding='utf-8') as file:
    content = file.read()

print(content)
```

```python
# Split into examples by detecting "# Example X:"
examples = re.split(r'# Example \d+:', content)
examples = [e.strip() for e in examples if e.strip()]

data = []
for idx, example in enumerate(examples, 1):
    query = re.search(r'<Query>(.*?)</Query>', example, re.DOTALL)
    document = re.search(r'<Document>(.*?)</Document>', example, re.DOTALL)
    answer = re.search(r'<Answer>(.*?)</Answer>', example, re.DOTALL)
    data.append({
        "Query": query.group(1).strip() if query else None,
        "Document": document.group(1).strip() if document else None,
        "Answer": answer.group(1).strip() if answer else None,
        "Context_Relevance_Label": "[[Yes]]",
        "Answer_Relevance_Label": "[[Yes]]",
        "Answer_Faithfulness_Label": "[[Yes]]"
    })
    
# Convert to DataFrame
df = pd.DataFrame(data)
df.to_csv("..\\datasets\\clinical_cases\\clinical_few_shot_prompt.tsv", sep="\t")
    
```
